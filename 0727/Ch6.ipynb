


# Filtering

```python
#average filter

source = cv2.imread("/content/drive/MyDrive/ABEP/Test1.BMP")
source = cv2.cvtColor(source, cv2.COLOR_BGR2GRAY)
cv2_imshow(source)
average = cv2.blur(source, ksize=(3,3))
cv2_imshow(average)

```

```python
#median filter
source = cv2.imread("/content/drive/MyDrive/ABEP/Test2.BMP")
source = cv2.cvtColor(source, cv2.COLOR_BGR2GRAY)
cv2_imshow(source)
result = cv2.medianBlur(source, 3)

cv2_imshow(result)
```

```python
#maximum filter
from scipy.ndimage import maximum_filter

source = cv2.imread("/content/drive/MyDrive/ABEP/Test2.BMP")
source = cv2.cvtColor(source, cv2.COLOR_BGR2GRAY)
cv2_imshow(source)
result = maximum_filter(source, 3)
cv2_imshow(result)
```

```python
#minimum filter
from scipy.ndimage import minimum_filter

source = cv2.imread("/content/drive/MyDrive/ABEP/Test3.BMP")
source = cv2.cvtColor(source, cv2.COLOR_BGR2GRAY)
cv2_imshow(source)
result = minimum_filter(source, 3)
cv2_imshow(result)
```

```python
#midpoint filter

source = cv2.imread("/content/drive/MyDrive/ABEP/Test4.bmp")
source = cv2.cvtColor(source, cv2.COLOR_RGB2GRAY)
cv2_imshow(source)

maxf = maximum_filter(source, 3)
minf = minimum_filter(source, 3)
midpoint = maxf/2 + minf/2
cv2_imshow(midpoint)
```




# Edge detection

```python
#sobel

source = cv2.imread("/content/drive/MyDrive/ABEP/Woman.bmp")
cv2_imshow(source)

# x-direction
img_sobel_x = cv2.Sobel(source,cv2.CV_64F,1,0,ksize=3)
img_sobel_x = cv2.convertScaleAbs(img_sobel_x)
cv2_imshow(img_sobel_x)

# y-direction
img_sobel_y = cv2.Sobel(source,cv2.CV_64F,0,1,ksize=3)
img_sobel_y = cv2.convertScaleAbs(img_sobel_y)
cv2_imshow(img_sobel_y)

#x+y-direction
img_sobel = cv2.addWeighted(img_sobel_x, 1, img_sobel_y, 1, 0)
cv2_imshow(img_sobel)
```

```python
#prewitt

source = cv2.imread("/content/drive/MyDrive/ABEP/Woman.bmp")
cv2_imshow(source)

# x-direction
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
img_prewitt_x = cv2.filter2D(source, -1, kernelx)
cv2_imshow(img_prewitt_x)

# y-direction
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
img_prewitt_y = cv2.filter2D(source, -1, kernely)
cv2_imshow(img_prewitt_y)

#x+y-direction
img_prewitt = cv2.addWeighted(img_prewitt_x, 1, img_prewitt_y, 1, 0)
cv2_imshow(img_prewitt)
```

```python
#roberts

source = cv2.imread("/content/drive/MyDrive/ABEP/Woman.bmp")
cv2_imshow(source)

# x-direction
kernelx = np.array([[1, 0], [0, -1]])
img_roberts_x = cv2.filter2D(source, -1, kernelx)
cv2_imshow(img_roberts_x)

# y-direction
kernely = np.array([[0, 1], [-1, 0]])
img_roberts_y = cv2.filter2D(source, -1, kernely)
cv2_imshow(img_roberts_y)

#x+y-direction
img_roberts = cv2.addWeighted(img_roberts_x, 1, img_roberts_y, 1, 0)
cv2_imshow(img_roberts)
```

```python
#laplacian

source = cv2.imread("/content/drive/MyDrive/ABEP/Woman.bmp")
cv2_imshow(source)
laplacian = cv2.Laplacian(source, cv2.CV_8U, ksize=3)
cv2_imshow(laplacian)
```

```python
#gaussian

source = cv2.imread("/content/drive/MyDrive/ABEP/Woman.bmp")
cv2_imshow(source)
gaussian = cv2.GaussianBlur(source, (0, 0), 1)
cv2_imshow(gaussian)
```

```python
#Laplacian of Gaussian

source = cv2.imread("/content/drive/MyDrive/ABEP/Woman.bmp")
cv2_imshow(source)
gaussian = cv2.GaussianBlur(source, (3, 3), 0)
log = cv2.Laplacian(gaussian, cv2.CV_64F, ksize=5)
cv2_imshow(log)
```

```python
#Difference of Gaussian (DoG)

source = cv2.imread("/content/drive/MyDrive/ABEP/Woman.bmp")
cv2_imshow(source)

low_sigma = cv2.GaussianBlur(source,(5,5),0)
high_sigma = cv2.GaussianBlur(source,(7,7),0)

dog = low_sigma - high_sigma
cv2_imshow(dog)
```




# Interpolation

```python
#nearest neighbor

source = cv2.imread("/content/drive/MyDrive/ABEP/small.bmp")
nearest = cv2.resize(source, None, fx=20, fy=20, interpolation = cv2.INTER_NEAREST)
cv2_imshow(source)
cv2_imshow(nearest)
```

```python
#bilinear 

source = cv2.imread("/content/drive/MyDrive/ABEP/small.bmp")
bilinear = cv2.resize(source, None, fx=20, fy=20, interpolation = cv2.INTER_LINEAR)
cv2_imshow(source)
cv2_imshow(bilinear)
```

```python
#bicubic

source = cv2.imread("/content/drive/MyDrive/ABEP/small.bmp")
bicubic = cv2.resize(source, None, fx=20, fy=20, interpolation = cv2.INTER_CUBIC)
cv2_imshow(source)
cv2_imshow(bicubic)
```
